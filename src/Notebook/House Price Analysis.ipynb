# Fun for deteting outlier

```python
def identify_outlier(col):
    # find 2 percentile
    q1=df[col].quantile(0.25)
    
    # find 75 percentile
    q3=df[col].quantile(0.75)
    
    # find IQR
    iqr=q3-q1
    
    # lower/min bound
    lower_bound=q1-1.5*iqr
    
    # upper/max bound
    upper_bound=q3+1.5*iqr
    
    # Find the outlier
    outlier=df[(df[col]<lower_bound) | (df[col]>upper_bound)]
    
    return outlier
```

```python
identify_outlier("price")
```

# Conclussion
**these are the outlier that will disturb the disturbution of data and we can't drop it b/c values are to large now we will turncate them**

# fun of finf the upper and lower bound

```python
def identify_bounds(col):
    # find 2 percentile
    q1=df[col].quantile(0.25)
    
    # find 75 percentile
    q3=df[col].quantile(0.75)
    
    # find IQR
    iqr=q3-q1
    
    # lower/min bound
    lower_bound=q1-1.5*iqr
    
    # upper/max bound
    upper_bound=q3+1.5*iqr
    
    return {"upper":upper_bound,"lower":lower_bound}
```

```python
identify_bounds("price")
```

# We will find the upper and lower bound now we will turncate them

```python
# np.where(conditon,true,false)
def turncate(col,upper,lower):
    df[col]=np.where(
    df[col]<lower,
        lower,
        
    np.where(df[col]>upper,
        upper,
             
    df[col]
    ))
    
    return df[col]
```

```python
turncate("price",1154353.57142825,-170589.28571375)
```

# Now we have a fun for identify outlier, upper and lower bound and also turncate fun now chain them

```python
def merge_bounds_turncate(col):
    # find 2 percentile
    q1=df[col].quantile(0.25)
    
    # find 75 percentile
    q3=df[col].quantile(0.75)
    
    # find IQR
    iqr=q3-q1
    
    # lower/min bound
    lower_bound=q1-1.5*iqr
    
    # upper/max bound
    upper_bound=q3+1.5*iqr
    
    #call the turncate fun
    return turncate(col,upper_bound,lower_bound)
```

```python
df["price"]=merge_bounds_turncate("price")
```

# Build a pipeline

### Num_pipeline

```python
num_transform=Pipeline(steps=[
    ("impute",SimpleImputer(strategy="mean")),
    ("num_transform",StandardScaler())
])
```

# cat-pipeline

```python
cat_transform=Pipeline(steps=[
    ("impute",SimpleImputer(strategy="most_frequent")),
    ("cat_transform",OneHotEncoder(handle_unknown="ignore"))
])
```

# Build a transformer

```python
process=ColumnTransformer(transformers=[
    ("num_transform",num_transform,num_col),
    ("cat_transform",cat_transform,cat_col)
])
```

```python
pipe=Pipeline([
    ("processor",process),
    ("model",LinearRegression())
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
```

# Make a function for evalution

```python
def evalute_error(actual,pred):
    mse=mean_squared_error(actual,pred)
    r2=r2_score(actual,pred)*100
    mas=mean_absolute_error(actual,pred)
    
    return {"mse":mse,"r2_score":r2,"mae":mas}
```

```python
mse=mean_squared_error(y_test,pred)
```

```python
evalute_error(y_test,pred)
```
